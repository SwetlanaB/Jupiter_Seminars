## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] =0
corr_matrix
```

```python
plt.figure(figsize =(25,10))
sns.heatmap(corr_matrix, annot =True, linewidths =5, cmap ='coolwarm');
```

 *Анализируя матрицу корреляции, можно отметить, что примерно половина характеристик имеют между собой линейную зависимость разной степени и направленности.* 
 ### Проанализируем зависимость с  прямой линейной зависимостью:
  - 0,9 зависимость наземной постройки от жилых кв.метров-весьма высокая;
  - 0,8 зависимость количества ванных комнат от количества жилой площади-высокая;
  - 0,5 зависимость оценки от этажей дома-заметная;  
  - 0,3 зависимость наличия подвала от вида из окон-слабая;
 ### Посмотрим на обратную зависимость с отрицательным коэффициентом корреляции:
  - -0,4 зависимость года постройки от состояния дома.


```python
sns.jointplot(x = df['sqft_above'], y = df['sqft_living'], kind ='reg');
```

*В построенном графике "Взаимосвязь наземной постройки от количества жилых метров " видна прямая зависимость двух факторов, они практически дублируют друг друга. Гипотеза логична:  построеные жилые квадратные метры практически и есть наземные строения, следовательно и изменяются они синхронно.*

```python
sns.jointplot(x = df['sqft_living'], y = df['bathrooms'], kind ='reg');
```

*В построенном графике "Взаимосвязь количества ванных комнат от количества жилых метров " есть прямая зависимость двух факторов, с ростом площади увеличевается количество ванных комнат: в левой части графика скопление точек в интервале от 0 до 6000 кв м пропорционально растет со  значениями ванных до шести .*

```python
sns.jointplot(x = df['floors'], y = df['grade'], kind ='kde');
```

*В построенном графике "Взаимосвязь оценки дома от количества этажей " видна прямая зависимость двух факторов. Она не сильная, составляет примерно 50%. Дома с одним и двумя этажами, их больше, их оценка изменяется в промежутке от 6 до 9(средний уровень конструкции и дизайна) значений и дает шесть ярковыраженных, плотно закрученных, с небольшим искривлением дисков. В этих  фигурах  прямая зависимость сильнее, чем в других разбросанных фигурах: следующий этаж имеет оценку выше( только в отношении 1-го и 2-го дома) *

```python
sns.jointplot(x = df['sqft_basement'], y = df['view'], kind ='kde');
```

*В построенном графике "Взаимосвязь наличия подвала от вида из окна " зависимость практически отсутствует, только для нулевого значения вида из окна наличие подвала, его площадь имеют значение. Соответственно, у домов с хорошим значением оценки вида нет влияния фактора наличия подвала.*

*В построенном графике "Взаимосвязь оценки состояний домов от этажности " существует обратная связь факторов: чем больше величина  оценки состояния дома, тем меньше у него этажей. Вероятно, .*

```python
plt.figure(figsize =(20,6))

sns.boxplot( x =df['condition'], y = df['yr_built'], whis =1.5)

plt.xlabel('condition')
plt.ylabel('yr_built')
plt.title('Взаимосвязь оценки состояния дома от года постройки');
```

*В построенном графике "Взаимосвязь года постройки от состояния дома " видна обратная зависимость  в правой части графика. Более позднему году постройки соответствует лучшее состояние дома.Т.к.коэффициент корреляции -0,4,  противовес ему составляет левая часть графика с прямой зависимостью в росте двух показателей до удовлетворительного значения.*
